```python
import random
def pieChart(axes,labels,sizes,explodes):
    explodes = [0.05 for i in range(len(labels))]
    color_list = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
             for i in range(len(labels)+1)]
    axes.pie(sizes,labels=labels,explode=explodes,colors=color_list[:-1], autopct='%1.1f%%',
            shadow=True, startangle=90)
    plt.tight_layout()
    return
```
